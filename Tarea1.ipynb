```python
#inicializamos algunas listas que nos ayudaran a guardar la informacion
dias=[]
articulos=[]
cuantos=[]
precio_ind=[]
```

Lo que se busca en el siguiente bloque es que el programa la pregunte a la persona si quiere generar las ventas de los ultimos dias, asi como registrar el articulo, la cantidad de ellos y su precio individual, esto por cada uno de los 5 dias, finalmente esta la opcion de salir y obtener tu data frame con la informacion administrada y con una columna más sobre el precio del articulo con descuento

```python
print("Bienvenido al menú interactivo")
while(True):
    print("""¿Qué quieres hacer? Escribe una opción
    1) Generar Ventas
    2) Salir""")
    opcion = input()
    if opcion == '1':
        print("""Hola, Vamos a generar las ventas de los ultimos 5 dias, Elige el dia
        1) Dia 1
        2) Dia 2
        3) Dia 3
        4) Dia 4
        5) Dia 5 """)
        dia= input()
        print("Escribe que articulo vendiste ")
        articulo=str(input())
        print("Escribe el numero de unidades por articulo que vendiste ")
        numero=int(input())
        print("Escribe sus precio")
        precio=int(input())

        # Ahora esta informacion la vamos a guardar en listas para que no se pierda
        dias.append(dia)
        articulos.append(articulo)
        cuantos.append(numero)
        precio_ind.append(precio)

    elif opcion =='2':
        print("¡Hasta luego! Ha sido un placer ayudarte, Aqui tienes tus datos en una tabla: ")
        break
    else:
        print("Comando desconocido, vuelve a intentarlo")
```

```python
df=pd.DataFrame()
df['Dia']=dias
df['Articulo']=articulos
df['Nro de Unidades']=cuantos
df['Precio']=precio_ind
df['Precio con Descuento']=df['Precio']*0.95
df
```

Vamos a crear una funcion que dado un diccionario con nombres y salario primero cree un data frame con dichos datos y luego cree un segundo df con los estadisticos requeridos como la media, el minimo y el maximo, asi como tambien pusimos la varianza y la sd

```python
def estadisticos(diccionario):
  df = pd.DataFrame([[key, diccionario[key]] for key in diccionario.keys()], columns=['Nombre', 'Salario Mensual'])
  dic={
      "Media": np.mean(df['Salario Mensual']),
      "varianza": np.var(df['Salario Mensual']),
      "Desviación Estándar": np.std(df['Salario Mensual']),
      "Mínimo": np.min(df['Salario Mensual']),
      "Máximo": np.max(df['Salario Mensual']),
      }
    
  df2=pd.DataFrame([[key, dic[key]] for key in dic.keys()], columns=['Estadistico', 'Valor'])
  return(df2)

estadisticos(clients)
```

Vamos a realizar un df con la informacion dada

```python

mes=['Enero', 'Febrero','Marzo','Abril','Mayo']
ventas=[30500,35600,28300,33900,42500]
gastos=[22000,23450,18100,35700,32450]
df2=pd.DataFrame()
df2['Mes']=mes
df2['Ventas']=ventas
df2['Gastos']=gastos
df2
```

```python
#Creamos la columna 1 que es el gasto/ las ventas
df2['Columna 1']=(df2['Gastos']/df2['Ventas'])*100
df2  
```

Para la siguiente columna crearemos un for, en el cual vamos a pasar la cantidad de ventas y vamos a preguntar si son superiores a 30000, en caso de ser cierto vamos asignar a una variable el valor "meta superada", en caso de que no, le asignaremos "meta no superada", cada valor se guarda en una lista para posteriormente esta lista unirla al df

```python
a=[]
for i in ventas:
  if i >30000:
    venta="Venta Superada"
  else:
    venta="Meta no Superada"
  a.append(venta)
df2['Columna 2']=a
df2
```

Para ver la diferencia vamos primero a convertir las listas de ventas y gastos a arrays para poder restarlas.
Una vez hecho esto si el resultado de restar los gastos a las ventas es negativo quiere decir que gastamos más de lo que se vendio y hay que obtener dicha diferencia.

Entonces creamos un for, donde si la diferncia es menor a 0(i.e. es negativa) va a multiplicar dicho valor por -1 y lo guardara en una variable,
si no es menor a 0, el resultado que guardara sera de 0.

Finalmente unimos esta lista al df

```python
venta_array=np.array(ventas)
gasto_array=np.array(gastos)
diferencia=venta_array-gasto_array
diferencia
b=[]

for i in list(diferencia):

  if i < 0:
    dif=i*-1
  else:
    dif=0

  b.append(dif)

df2['Columna 3']=b
df2
```
